```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def read_data(file):
    data = []
    with open(file) as fd:
        for l in fd:
            data.append(l.split())
    return data

def extract_rel_stats(data):
    info = {}
    for _,r,_ in data:
        if r not in info:
            info[r] = 0
        info[r]+=1
    return info
```

```python
# Ploting data
data = read_data('real-data')
info = extract_rel_stats(data)
print(info)
rels = list(info.keys())
rels.sort()
plt.title('Tripletas por relaciones')
for i,key in enumerate(rels):
    plt.bar(i, info[key])
plt.xticks(np.arange(len(info))+0.01, rels)
plt.savefig('real-data.png')
```

```python
# Ploting noise
data = read_data('noise')
info = extract_rel_stats(data)
rels = list(info.keys())
rels.sort()
for i,key in enumerate(rels):
    plt.bar(i, info[key])
plt.xticks(np.arange(len(info))+0.01, rels)
plt.savefig('noise-data.png')
```
