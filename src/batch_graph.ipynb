```python
import matplotlib.pyplot as plt
import matplotlib.mlab as mlb
import numpy as np
```

```python
data = np.genfromtxt('charts/batch.logs.csv', delimiter=',')
```

```python
X = np.arange(len(data))
L = data[:,1]
A = data[:,2]
```

```python
DEG=4
pol_L = np.polyfit(X, L, deg=DEG)
pol_A = np.polyfit(X, A, deg=DEG)

L2 = np.polyval(pol_L, X)
A2 = np.polyval(pol_A, X)
```

```python
plt.plot(X,L2, label='loss')
plt.plot(X,A2, label='acc')
plt.title("Comportamiento durante el entrenamiento")
plt.legend()
plt.savefig('training.png')
```
